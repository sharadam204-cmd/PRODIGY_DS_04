```
import pandas as pd
df = pd.read_csv('twitter_training.csv', header=None, names=['ID', 'Topic', 'Sentiment', 'Tweet'])
```

```
df.head()
df.columns
```

```
df['Sentiment'].value_counts()
```

```
import matplotlib.pyplot as plt
df['Sentiment'].value_counts().plot(kind='bar')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```

df['Sentiment'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Sentiment Analysis')
plt.ylabel('')
plt.show()
```

```
import pandas as pd
val_df = pd.read_csv('twitter_validation.csv', header=None, names=['ID', 'Topic', 'Sentiment', 'Tweet'])
val_df.head()
```

```
val_df.shape
```

```
val_df['Sentiment'].value_counts()
```

```
import matplotlib.pyplot as plt
val_df['Sentiment'].value_counts().plot(kind='bar')
plt.title('Validation Dataset Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```
val_df['Sentiment'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Validation Sentiment Analysis')
plt.ylabel('')
plt.show()
```

```
train_counts = df['Sentiment'].value_counts()
val_counts = val_df['Sentiment'].value_counts()

comparison = pd.DataFrame({
    'Training': train_counts,
    'Validation': val_counts
})
comparison.plot(kind='bar')
plt.title('Training vs Validation Sentiment Comparison')
plt.ylabel('Count')
plt.show()
```
